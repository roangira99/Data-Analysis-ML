```python
from scipy import stats
# Creating a variable to examine a feature
# var = df['Age (day)']
# var = df['Cement (component 1)(kg in a m^3 mixture)']
# var2 = df['Concrete compressive strength(MPa, megapascals) ']


df_regression = df[['Concrete compressive strength(MPa, megapascals) ', 
                           'Cement (component 1)(kg in a m^3 mixture)']].copy()

# Get the regression slope and intercept
m, b, r, p, se = stats.linregress(df_regression['Cement (component 1)(kg in a m^3 mixture)'], 
                                df_regression['Concrete compressive strength(MPa, megapascals) '])
print('slope: {:.4f}\ny-intercept: {:.4f}'.format(m,b))
print('so.../n f(x) = {:.4f}x + {:.4f}'.format(m,b))

# Using the function (mx + b) to calculate f(x) for each x(Cement) value
df_regression['fx'] = (m * df_regression['Cement (component 1)(kg in a m^3 mixture)'] + b)

# Calculate the error between f(x) and the actual 
#y (Concrete compressive strength) value
df_regression['error'] = df_regression['fx'] - df_regression['Concrete compressive strength(MPa, megapascals) ']

# # Creating a figure
# fig = plt.figure(figsize=(10, 4))

# Plotting a scatter plot
df_regression.plot.scatter(x='Cement (component 1)(kg in a m^3 mixture)', y='Concrete compressive strength(MPa, megapascals) ')

# Plotting the regression line
plt.plot(df_regression['Cement (component 1)(kg in a m^3 mixture)'], df_regression['fx'], color='red')

plt.show()

# plt.scatter(var, var2)
```

<b>The next step will be to split the dataset into training and testing sets, then use the k-nearest neighbors regressor algithm to train a model using the sets.</b>

```python
X = df.drop('Concrete compressive strength(MPa, megapascals) ', axis=1) # Dropping the target column to remain with features only
X = X.values
y = df['Concrete compressive strength(MPa, megapascals) '] # creating a variable for the target/label
y = y.values
```

```python
# Printing the feature columns
print(X)
```

```python
print(y)
```

```python
# Splitting the model into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12345)
```

```python
# Training the model
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor(n_neighbors=3)
knn_model.fit(X_train, y_train)
```

<b>Predicting for training data and calculating the prediction error. </b>

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
train_preds = knn_model.predict(X_train)
mse = mean_squared_error(y_train, train_preds)
rmse = sqrt(mse)
print(rmse)
```

<b>Predict for test data and calculate prediction error<br>
If lower then the model can generalise well. <br>
If higher then the model is overfitting.</b>

```python
test_preds = knn_model.predict(X_test)
mse = mean_squared_error(y_test, test_preds)
rmse = sqrt(mse)
print(rmse)
```

By observing the root mean square errors on the training and test data, it is evident that the error rate of the model is lower on the training data than on the test data. This means that it is <b>overfitting</b>.<br> To improve the model's performance, we can change the number of neighbors. Using <b>grid search</b>, we can try several different values of k and then choose the best k.


### Grid Search

```python
from sklearn.model_selection import GridSearchCV
parameters = {"n_neighbors": range(1, 50)}
gridsearch = GridSearchCV(KNeighborsRegressor(), parameters)
gridsearch.fit(X_train, y_train) # Train using the best k found
```

```python
gridsearch.best_params_
```

```python
train_preds_grid = gridsearch.predict(X_train)
train_mse = mean_squared_error(y_train, train_preds_grid)
train_rmse = sqrt(train_mse)                                     

test_preds_grid = gridsearch.predict(X_test)
test_mse = mean_squared_error(y_test, test_preds_grid)
test_rmse = sqrt(test_mse)
print("Train error:", train_rmse)
print("Test error:", test_rmse)
```

After performing hyperparameter tuning using grid search, we still get <b>n_neighbors = 3 </b> as the best <b>k</b> value. We can therefore conclude that the model's optimal performance on the current state of the dataset is achieved by setting he number of neighbors to 3.
